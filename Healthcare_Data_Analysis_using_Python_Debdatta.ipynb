### Data transformation

Name column values in proper format

```python
df['Name']= df['Name'].str.title()
```

Date columns convert in date format

```python
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'], format="%d-%m-%Y")
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'], format="%d-%m-%Y")
```

New column created

```python
# "Length of stay" column created

df["Length of stay"] = (df['Discharge Date'] - df['Date of Admission']).dt.days
```

```python
# Age group column created

def age_gr(age):
  if age > 75:
    return "75+"
  elif age >= 60:
    return "60-75"
  elif age >= 45:
    return "45-59"
  elif age >= 30:
    return "30-44"
  elif age >= 18:
    return "18-29"
  else:
    return "0-17"

df["Age group"] = df['Age'].apply(age_gr)
```

Exploratory Data Analysis (EDA)

### Patient Demographics

```python
# Which age group has the highest number of patients?

no_of_patients_by_age_group = df.groupby("Age group")["Name"].count().sort_values(ascending=False)

no_of_patients_by_age_group.plot(kind= "barh")
plt.title("Patients by Age group")
plt.ylabel("")
plt.show()
```

```python
# Gender distribution across different age groups.

pivot1 = pd.pivot_table(df, index='Age group', columns='Gender', values='Name', aggfunc='count')

pivot1.plot(kind="bar")
plt.title("Age group wise Gender distribution")
plt.xlabel("")
plt.show()
```

```python
# Patient count by medical condition.

patients_by_medical_codition = df.groupby("Medical Condition")["Name"].count().sort_values(ascending=False)
patients_by_medical_codition
```

### Admission & Hospital Operations

```python
# Admission type distribution (Emergency / Urgent / Elective).

admission_distribution = df["Admission Type"].value_counts()
admission_distribution
```

```python
df['Admission Type'].value_counts().plot(kind='bar', rot=0)
plt.title("Admission Type Distribution")
plt.xlabel("Admission Type")
plt.ylabel("Number of Admissions")
plt.show()
```

```python
# Which admission type has the longest average stay?

admission_type_avg_stay = df.groupby("Admission Type")["Length of stay"].mean().sort_values(ascending= False).round(0)
admission_type_avg_stay
```

```python
# Which hospitals admit the most patients?
top_hospitals = df["Hospital"].value_counts().head(10)
top_hospitals
```

```python
df['Hospital'].value_counts().head(10).plot(kind='barh')
plt.title("Top 10 Hospitals by Admissions")
plt.xlabel("Patient Count")
plt.show()
```

```python
 # Monthly trend of hospital admissions.

df["Admission_month"] = df['Date of Admission'].dt.to_period("M")

monthly_admission = df.groupby("Admission_month").size()
monthly_admission.plot(marker='o', figsize=(15,5))
plt.title("Monthly Trend of Hospital Admissions")
plt.xlabel("Month")
plt.ylabel("Number of Admissions")
plt.show()
```

### Analysis Summary

**Patient Demographics:**
*   The age group `60-75` has the highest number of patients, followed by `45-59` and `30-44`. The `0-17` age group has the fewest patients.
*   The gender distribution across age groups is relatively balanced, with slight variations.
*   Among medical conditions, `Arthritis`, `Diabetes`, and `Hypertension` have the highest patient counts, all closely distributed.

**Admission & Hospital Operations:**
*   Admission types (`Elective`, `Urgent`, `Emergency`) are fairly evenly distributed.
*   The average length of stay is similar across all admission types, with `Emergency` and `Elective` slightly higher at approximately 16 days, and `Urgent` at 15 days.
*   `LLC Smith`, `Ltd Smith`, and `Johnson PLC` are among the hospitals that admit the most patients.
*   Monthly hospital admissions show fluctuations but no strong overall increasing or decreasing trend over the observed period.

**Billing & Cost Analysis:**
*   The average billing amount is approximately `25539.32`, with a maximum of `52764.28` and a minimum of `-2008.49`.
*   `Obesity` shows the highest average billing amount, closely followed by `Diabetes` and `Asthma`.
*   Billing amounts are consistent across different admission types.
*   `Medicare` and `Blue Cross` show slightly higher average billing amounts among insurance providers.
*   The top billing patients were identified.

**Length of Stay Analysis:**
*   The overall average length of stay is approximately `16` days.
*   `Asthma` patients have a slightly longer average stay, while `Diabetes` patients have a slightly shorter average stay compared to other medical conditions.
*   Length of stay is similar across all admission types and age groups, generally hovering around 15-16 days.

**Test Results & Health Outcomes:**
*   Test results are distributed relatively evenly among `Abnormal`, `Normal`, and `Inconclusive` categories.
*   `Arthritis` and `Diabetes` are the medical conditions with the most abnormal test results.
*   There is no significant relationship between the length of stay and test results, as the average length of stay is similar across all test result categories.
*   `Lipitor`, `Ibuprofen`, and `Aspirin` are the most frequently prescribed medications.
